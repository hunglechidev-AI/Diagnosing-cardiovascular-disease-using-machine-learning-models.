# Models

```python
X1 = dia.drop('Diabetes',axis=1)
y1 = dia[['Diabetes']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score , classification_report, ConfusionMatrixDisplay,precision_score,recall_score, f1_score,roc_auc_score,roc_curve

X_train1 , X_test1 , y_train1 , y_test1 = train_test_split(X1 , y1 , test_size=0.2,random_state=42)
```

```python
models={
    "Logisitic Regression" :LogisticRegression(max_iter=20000),
    "Decision Tree" :DecisionTreeClassifier(),
    "Random Forest":RandomForestClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=3),
    "Multinomial Naive Bayes": MultinomialNB()
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train1,y_train1.values.ravel()) # Train Model
    # Make predictions
    y_train_pred = model.predict(X_train1)
    y_test_pred =  model.predict(X_test1)

  # Test set performance
    model_test_accuracy = accuracy_score(y_test1, y_test_pred)
    model_test_f1 = f1_score(y_test1, y_test_pred, average='weighted')
    model_test_precision = precision_score(y_test1, y_test_pred , average='weighted')
    model_test_recall  = recall_score(y_test1, y_test_pred,average='weighted')

  # Training set performance
    model_train_accuracy = accuracy_score(y_train1, y_train_pred)
    model_train_f1 = f1_score(y_train1, y_train_pred, average= 'weighted')
    model_train_precision = precision_score(y_train1, y_train_pred,average='weighted')
    model_train_recall = recall_score(y_train1, y_train_pred,average='weighted')

    print(list(models.keys())[i])

    print('Model performance for Training set (Diabetes)')
    print("- Accuracy: {:.4f}".format(model_train_accuracy))
    print('- F1 score: {:4f}'.format(model_train_f1))
    print('- Precision: {:4f}'.format(model_train_precision))
    print('- Recall: {:4f}'.format(model_train_recall))

    print('----------------------------------')

    print('Model performance for Test set (Diabetes)')
    print('- Accuracy: {:.4f}'.format(model_test_accuracy) )
    print('- Fl score: {:.4f}'.format(model_test_f1))
    print('- Precision: {:.4f}'.format(model_test_precision))
    print('- Recall: {:.4f}'.format(model_test_recall))


    print('='*35)
    print('\n')
```
